```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('./titanic.csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
mean_age = df['Age'].mean()
median_fare = df['Fare'].median()
mode_class = df['Pclass'].mode()[0]
```

```python
print(f'Mean Age: {mean_age}, Median Fare: {median_fare}, Mode Ticket Class: {mode_class}')
```

```python
contingency_table = pd.crosstab(df['Sex'], df['Survived'])
```

```python
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f'Chi-Square Statistic: {chi2}, p-value: {p}')
```

```python
model = ols('Fare ~ C(Pclass)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

regression_model = sm.Logit(df['Survived'], df[['Age', 'Sex', 'Pclass']].fillna(0)).fit()
print(regression_model.summary())
```
